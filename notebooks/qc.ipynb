```python
import numpy as np
import tifffile as tiff
from scipy import ndimage as ndi

def qc(mask, I, t1, t2):
    H, W = mask.shape[-2], mask.shape[-1]
    fg_frac = float(mask.mean())
    labeled, n = ndi.label(mask, structure=ndi.generate_binary_structure(3,1))
    comp_sizes = np.bincount(labeled.ravel())
    largest = int(comp_sizes[1:].max()) if comp_sizes.size > 1 else 0
    largest_frac = largest / mask.size
    mid = (I >= t1) & (I < t2)
    mid_claim_fg = (mask & mid).sum() / max(1, mid.sum())

    return {
        "fg_frac": fg_frac,
        "n_components": int(n),
        "largest_frac": largest_frac,
        "mid_claim_fg": float(mid_claim_fg),
    }

litarion = tiff.imread("../results/Litarion/Litarion_mask.tif")
original_litarion = tiff.imread("../data/Litarion.tif")
qc(litarion, original_litarion, 0.33003193, 0.5407566)
```
